6. Explain the concept of generators in Python and how they are defined.

Generators are special types of iterators that can be used to generate a sequence of values on the fly, rather than storing them in memory all at once.

Example:

def infinite_sequence():

    num = 0
    while True:
        yield num
        num += 1

sequence = infinite_sequence()

print(next(sequence))

0

print(next(sequence))

1

print(next(sequence))

2



7.What are the advantages of using generators over regular functions?

- Generators have several advantages over regular functions, including:

Memory efficiency: Generators only store the current state of the iteration, rather than storing all values in memory at once.

 Flexibility: Generators can be used to implement cooperative multitasking, where multiple tasks yield control to each other at specific points.

Improved performance: Generators can be faster than regular functions because they avoid the overhead of creating and returning lists.

Applies a function to each item in an iterable

Returns a new iterable with the results

Purpose: Transform data

Applies a function to all items in an iterable, cumulatively

Returns a single output value

Purpose: Aggregate data

--Filter()

Applies a predicate function to each item in an iterable

Returns a new iterable with only the items for which the predicate is True

Purpose: Select data

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

numbers = [1, 2, 3, 4, 5, 6]
print(sum_even_numbers(numbers))  # Output: 12

```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

for i in range(1, 201):
    if is_prime(i):
        print(i)


```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.n > 0:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.n -= 1
            return result
        else:
            raise StopIteration

fib = Fibonacci(10)
for num in fib:
    print(num)
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

for power in powers_of_2(5):
    print(power)

```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(filename):
    """
    Generator function to read a file line by line.

    Args:
        filename (str): The name of the file to read.

    Yields:
        str: Each line in the file as a string.
    """
    try:
        with open(filename, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"File '{filename}' not found.")


filename = 'example.txt'
for line in read_file_line_by_line(filename):
    print(line)

```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fa#hrenheit.

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

temperatures = [0, 10, 20, 30, 40]
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, temperatures))
print(fahrenheit_temperatures)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join([char for char in s if char not in vowels])

s = "Hello World"
print(remove_vowels(s))
```

```python
# 11. Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- € if the value of the order is smaller than 100,00 €.

# Write a Python program using lambda and map.


order_numbers = [34587, 98762, 77226, 88112]
book_titles = ['Learning Python, Mark Lutz', 'Programming Python, Mark Lutz', 'Head First Python, Paul Barry', 'Einführung in Python3, Bernd Klein']
quantities = [4, 5, 3, 3]
prices = [40.95, 56.80, 32.95, 24.99]


def calculate_order(order_number, quantity, price):
    total = quantity * price
    if total < 100:
        total += 10  # Add 10€ if total is less than 100€
    return (order_number, total)

# Use of map with lambda function to calculate the results

order_totals = list(map(lambda data: calculate_order(data[0], data[1], data[2]), zip(order_numbers, quantities, prices)))

print(order_totals)
```
